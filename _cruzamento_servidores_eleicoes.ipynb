```python
import pandas as pd
from unidecode import unidecode
```

### 1. Dados do SIAPE

```python
lic_2020_03_11 = pd.read_csv('AFASTREM_032020.csv', sep = ';', encoding = 'latin1')
```

```python
for i in ['AFASTREM_042020.csv', 'AFASTREM_052020.csv', 'AFASTREM_062020.csv', 'AFASTREM_072020.csv', 'AFASTREM_082020.csv', 'AFASTREM_092020.csv', 'AFASTREM_102020.csv', 'AFASTREM_112020.csv']:
    lic_2020_03_11 = pd.concat([lic_2020_03_11, pd.read_csv(i, sep = ';', encoding = 'latin1')])
    print(i)
```

```python
lic_2020_03_11[lic_2020_03_11['Descrição do afastamento'] == 'Lic. Atividade Politica Com Remuneração - EST               '][['Descrição do afastamento', 'Ano/Mês início afastamento', 'CPF']].groupby(['Descrição do afastamento', 'Ano/Mês início afastamento']).nunique()
```

```python
def corrige_texto(x):
    x_ = x.replace('  ', '')
    if x_[-1] == ' ':
        y = x_[:-1]
    else:
        y = x_
    return y

lic_2020_03_11['Nome'] = lic_2020_03_11['Nome'].map(corrige_texto)
```

```python
lic_2020_03_11['cpf_miolo'] = lic_2020_03_11['CPF'].apply(lambda x: x[3:-2])
```

```python
lic_2020_03_11.head(3)
```

```python
df_alvos = lic_2020_03_11[(lic_2020_03_11['Descrição do afastamento'] == 'Lic. Atividade Politica Com Remuneração - EST               ') & (lic_2020_03_11['Ano/Mês início afastamento'] >= 202008)]
```

```python
df_alvos.head()
```

```python
list(df_alvos)
```

```python
def corrige_numero(x):
    y = float(x.replace(',', '.'))
    return y

df_alvos['Valor rendimento liquido'] = df_alvos['Valor rendimento liquido'].map(corrige_numero)
```

```python
df_alvos_resumo = df_alvos[['Nome', 'CPF', 'cpf_miolo', 'UF da UPAG de vinculação', 'Ano/Mês início afastamento']].groupby(['Nome', 'CPF', 'cpf_miolo', 'UF da UPAG de vinculação']).count().reset_index()\
                 .merge(df_alvos[['Nome', 'CPF', 'cpf_miolo', 'UF da UPAG de vinculação', 'Valor rendimento liquido']].groupby(['Nome', 'CPF', 'cpf_miolo', 'UF da UPAG de vinculação']).mean().reset_index()\
                 , how = 'inner', on = ['Nome', 'CPF', 'cpf_miolo', 'UF da UPAG de vinculação'])

df_alvos_resumo.columns = ['Nome', 'CPF', 'cpf_miolo', 'UF da UPAG de vinculação', 'num_meses_registro_afastamento', 'Valor rendimento liquido']
```

### 2. Dados do TSE

```python
df_tse_votos = pd.read_csv('../dados_tse/2020/votacao_candidato_munzona_2020_BRASIL.csv', sep = ';', encoding = 'latin1')
df_tse_votos_grupo = df_tse_votos[['SG_UF', 'DS_CARGO', 'SQ_CANDIDATO', 'NM_CANDIDATO', 'SG_PARTIDO', 'DS_SIT_TOT_TURNO', 'DS_SITUACAO_CANDIDATURA', 'QT_VOTOS_NOMINAIS']].groupby(['SG_UF', 'DS_CARGO', 'SQ_CANDIDATO', 'NM_CANDIDATO', 'SG_PARTIDO', 'DS_SIT_TOT_TURNO', 'DS_SITUACAO_CANDIDATURA']).sum().reset_index()
```

```python
df_tse_votos_grupo = df_tse_votos_grupo.merge(df_candidatos[['SQ_CANDIDATO', 'NR_CPF_CANDIDATO', 'ST_CANDIDATO_INSERIDO_URNA']], on = 'SQ_CANDIDATO', how = 'inner')
```

```python
df_tse_votos_grupo
```

```python
def corrige_cpf(x):
    y = '0'*(11-len(str(x)))+str(x)
    return y

df_tse_votos_grupo['NR_CPF_CANDIDATO_ajustado'] = df_tse_votos_grupo['NR_CPF_CANDIDATO'].map(corrige_cpf)
```

```python
df_tse_votos_grupo['cpf_miolo'] = df_tse_votos_grupo['NR_CPF_CANDIDATO_ajustado'].apply(lambda x: x[3:-2])
```

```python
def corrige_nome(x):
    y = unidecode(x)
    return y

df_tse_votos_grupo['NM_CANDIDATO'] = df_tse_votos_grupo['NM_CANDIDATO'].map(corrige_nome)
```

### 3. Cruzamento

#### 3.1. Tem registro de afastamento mas não está na base de candidatos

```python
df_alvos_votos_t1 = df_alvos_resumo.merge(df_tse_votos_grupo[['cpf_miolo', 'NM_CANDIDATO', 'DS_SITUACAO_CANDIDATURA', 'QT_VOTOS_NOMINAIS']], how = 'left', left_on = ['cpf_miolo', 'Nome'], right_on = ['cpf_miolo', 'NM_CANDIDATO'])
```

```python
df_resultado_t1 = df_alvos_votos_t1[pd.isna(df_alvos_votos_t1['NM_CANDIDATO']) == True]
```

```python
df_resultado_t1.head()
```

```python
len(df_resultado_t1)
```

#### 3.2. Tem registro mas tem pouco voto (menos que 50)

```python
df_alvos_votos_t2 = df_alvos_resumo.merge(df_tse_votos_grupo[['cpf_miolo', 'NM_CANDIDATO', 'DS_SITUACAO_CANDIDATURA', 'ST_CANDIDATO_INSERIDO_URNA', 'QT_VOTOS_NOMINAIS']], how = 'inner', left_on = ['cpf_miolo', 'Nome'], right_on = ['cpf_miolo', 'NM_CANDIDATO'])
```

```python
df_resultado_t2 = df_alvos_votos_t2[(df_alvos_votos_t2['DS_SITUACAO_CANDIDATURA'] == 'APTO') & (df_alvos_votos_t2['QT_VOTOS_NOMINAIS'] < 50) & (df_alvos_votos_t2['ST_CANDIDATO_INSERIDO_URNA'] == 'SIM')].sort_values('QT_VOTOS_NOMINAIS', ascending = True)
```
